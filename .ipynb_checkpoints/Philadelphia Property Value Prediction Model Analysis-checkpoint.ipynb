# Philadelphia Property Value Prediction Model Analysis 
---
## Introduction

Permits are required in most jurisdictions whenever developers or homeowners want to alter or remodel their homes, construct new buildings, and/or demolish existing structures.  The type of permit required depends on both the local jurisdiction and the scope of work performed. In Philadelphia, Pennsylvania, permits are issued by The Department of Licenses & Inspections and the data for those permits are publicly available at https://www.opendataphilly.org/.  For this analysis, two files downloaded from [the Licenses and Inspections Building and Zoning Permits page](https://www.opendataphilly.org/dataset/licenses-and-inspections-building-permits) of this website will be used, listing information for permits from 2007 to the present.  

The other data to be utilized in this analysis will be housing price data made available by Zillow at https://www.zillow.com/research/data/.  The data from Zillow is based on the Zillow Home Value Index (ZHVI) when represents the “typical home value for the region” as defined in Zillow's [methodology guide](https://www.zillow.com/research/zhvi-methodology-2019-highlights-26221/).  

The goal of this analysis is to determine whether the number of permits issued in a given zip code within prior years is correlated to property value and changes of property values within the zip code.  The assumption going in is that building permits that better the buildings within a zip code should have some correlation with the future housing prices and change in housing prices.  Thus the theory is that if we're designing a machine learning model, the inclusion of permits would improve the model. 

The limitations for this analysis include the following:
- The permit data available ranges back from 2007, which only gives 15 years of analysis.  
- The housing market crash of 2008 occured during the short timeframe available within our data.  This crash would be assumed to have affected housing price data and may overshadow any correlations between housing prices and permit issuance. 
- Not all renovation work on a building requires a permit, which would limit the data available within our permit dataset.  In addition, some work is occasionally performed without proper permitting, which would again limit data available in our dataset.

With all of that understood, we can begin exploring the data and performing the analysis.

### Zillow Data

First we'll read the Zillow Home Data into a DataFrame.

Next we'll remove any of the rows with locations that are not Philadelphia, Pennsylvania

```python
# Remove data not in Philadelphia, Pennsylvania
zillow = zillow[(zillow['City'] == 'Philadelphia') & (zillow['State'] == 'PA')]
zillow.head()
```

To account for any errors in data entry for the city and state, we'll also check the zipcodes column.  We'll remove any rows with zipcodes not within the list of valid Philadelphia, PA zipcodes.  A list of zipcodes within Philadelphia, PA was obtained from https://www.zillow.com/browse/homes/pa/philadelphia-county/ and is replicated below.

Only keep rows from data with a zipcode within the list above.

```python
# rename RegionName columns and cast as strings
zillow.rename(columns={'RegionName': 'zipcode'}, inplace=True)
zillow['zipcode'] = zillow['zipcode'].astype('str')
zillow = zillow[zillow['zipcode'].isin(list_of_zipcodes)]
zillow.head()
```

We're only going to look at house values year over year rather than each month in order to smooth out fluctuations like seasonal changes, etc. that affect number of permits (due to weather constraints) and housing values.  We're interested in year-over-year changes within this analysis, not seasonal ones, so we'll need to simplify the columns in the data by taking the mean of the 12 months of the year.  We're also dropping the year 2022 from the data since the year is ongoing and thus it would be partial data.

We'll keep those yearly housing value columns along with the 'zipcode' column, and drop all others as they're not relevant to this analysis.

```python
groupby_dict = {}
year_list = [str(x) for x in range(2000,2022)]
for col in zillow.columns:
    for year in year_list:
        if year in col:
            groupby_dict[col] = year
groupby_dict['zipcode'] = 'zipcode'

zillow = zillow.groupby(groupby_dict, axis = 1).mean()
zillow.head()
```

We're going to alter the DataFrame so that the only columns are 'zipcode', 'year', and 'hv (housing value).  This will allow us to compute percent changes for previous years and will help us to merge this data with the permit data.

```python
zillow = zillow.melt(id_vars=["zipcode"], var_name="year", value_name="hv")
zillow['zipcode'] = zillow['zipcode'].astype('int').astype('str')
zillow['year'] = zillow['year'].astype(int)
zillow.head()
```

Now we'll create columns showing house values (`hv_*yb`) from previous years (1, 3, and 5 years ago) along with columns showing the percent change (`pc_*yb`) from previous years (1, 3, and 5 years ago) along each zipcode in a given year.  We'll also create columns that show the house value one year in the future (`hv_1yf`) and percent change one year in the future(`pc_1yf`), utilizing the data within the DataFrame.

```python
## First define a groupby function:
def prev_prices(data):
    data = data.sort_values("year", ascending=True)
    
    # housing values 1 year, 3 years, and 5 years prior
    data["hv_1yb"] = data["hv"].shift(1)
    data["hv_3yb"] = data["hv"].shift(3)
    data["hv_5yb"] = data["hv"].shift(5)
    
    # percent change since 1 year, 3 years, and 5 years prior
    data["pc_1yb"] = (data["hv"] - data["hv_1yb"]) / data["hv_1yb"] * 100
    data["pc_3yb"] = (data["hv"] - data["hv_3yb"]) / data["hv_3yb"] * 100
    data["pc_5yb"] = (data["hv"] - data["hv_5yb"]) / data["hv_5yb"] * 100
    
    # housing values 1 year in the future
    data["hv_1yf"] = data["hv"].shift(-1)
    
    # percent change 1 year in the future
    data["pc_1yf"] = (data["hv"] - data["hv_1yf"]) / data["hv_1yf"] * 100

    return data

## Groupby and apply
zillow = zillow.groupby(["zipcode"]).apply(prev_prices)
```

## Data Analysis
### Setup
Let's look at our two dataframes:

```python
zillow.head()
```

Before combining these dataframes, we'll be calculating the number of permits grouped by both zipcode and year.  However, which permit types to include is still a question.  Let's take a look at the `permitdescription` column.

```python
permits_combined['permitdescription'].value_counts()
```

```python
permit_type_list = permits_combined['permitdescription'].unique()
len(permit_type_list)
```

There are 25 unique permit types within the 'permitdescription' column.  Some of the ones that stand out as far as relevance for affecting housing values would be the following:
- NEW CONSTRUCTION PERMIT
- ZONING PERMIT
- ALTERATION PERMIT
- RESIDENTIAL BUILDING PERMIT
- COMMERCIAL BUILDING PERMIT
- ADDITION PERMIT
- DEMOLITION PERMIT

However, rather that just selecting the ones that sound obvious, we're going to do the following:
- Loop through each permit type
- Compute the number of permits per zipcode and year which will include creating the following columns:
   - `no_permits` - number of permits in the given zipcode and year
   - `np_1yb` - number of permits in the prior year
   - `np_3yb` - number of permits in the prior 3 years
   - `np_5yb` - number of permits in the prior 5 years
- Combine the permit data and the Zillow data
- Analyze correlation between the permit columns (shown above) VS the housing value (`hv`) and percent change columns (`pc_*`)

Let's start by creating functions.

```python
# Compute rolling permit data for 1y, 3y, and 5y ago
def prev_permits(data):
    data = data.sort_values("year", ascending=True)
    data["np_1yb"] = data["no_permits"].shift(1)
    data["np_3yb"] = data.rolling(3, closed="left", min_periods=3).sum()["no_permits"]
    data["np_5yb"] = data.rolling(5, closed="left", min_periods=5).sum()["no_permits"]
    return data

# Compute correlation using permit_type as 'all' if not specified
def created_combined_df(permit_data, permit_type='all'):
    
    # Isolate the data to the rows related to the entered permit_type 
    if permit_type != 'all':
        permit_data = permit_data[permit_data['permitdescription'].isin(permit_type)]

    # Count number of permits by zipcode and year
    permit_data = permit_data.groupby(['zipcode','year'])['zipcode'].count().reset_index(name='no_permits')

    # Apply groupby function to add rolling sums to table
    permit_data = permit_data.groupby(["zipcode"]).apply(prev_permits)

    # Combine permit and housing data on zipcode and year
    combined_data = pd.merge(permit_data, zillow, on=['zipcode','year'])
    
    # Drop rows with null values
    combined_data = combined_data.dropna()
    
    return combined_data
```

### All Permit Types

First, before looping, we'll show an example of the correlation of the data using 'all' permit types.  The final table after running the function has the following columns:
- `zipcode`: zipcode
- `year`: year
- `no_permits`: number of permits issued within a given year and zipcode
- `np_1yb`, `np_3yb`, `np_5yb`: number of permits in the past year, 3 years, and 5 years respectively (rolling sums).
- `hv`: average home value within a given year and zipcode
- `hv_1yb`, `hv_3yb`, `hv_5yb`: average home value within a zipcode 1, 3, and 5 years prior to the year within the respective row.
- `pc_1yb`, `pc_3yb`, `pc_5yb`: percentage change in average home value in the past year, past 3 years, and past 5 years, respectively.
- `hv_1yf`: average home value within a zipcode 1 year into the future of the year within the respective row.
- `pc_1yf`: percentage change in average home value in the next year.

```python
housing_data = created_combined_df(permits_combined)
housing_data.head(20)
```

```python
sns.heatmap(housing_data.corr().round(2), annot = True)
plt.title("Correlation Matrix for Housing Data (All Permit Types)")
plt.show()
```

This is the full correlation matrix, however our focus is utilizing historic columns to predict columns in the future.  We can alter the heatmap to show only those relationships, per below.

```python
sns.heatmap(housing_data.corr().iloc[1:12, 12:], annot = True)
plt.title("Correlation Matrix for Housing Data (All Permit Types)")
plt.xlabel("House Value Columns (Future)")
plt.ylabel("Permit Columns & House Value Columns (Past & Current)")
plt.show()
```

Notable from this is the moderate to strong correlations between the `pc_1yf` column looking one year forward, and the three `pc_*yb` columns looking backwards in the bottom righthand corner.  The permit columns (`no_permits` and `np_*yb`) appear to have a weak negative correlation with the `pc_1yf` column, and a very weak correlation with the `hv_1yf` column. Historic home values (`hv_*yb`) also have a very strong correlation with future housing values `hv_1yf`, which makes sense. 

We're going to now loop through each of the unique values in the `permitdescription` column and compute the correlation between the Permit Columns and the House Value Columns, as we did above.  Instead of producing a heatmap for each of the 25 values, we'll display the max correlation coefficient in order to select relevant permits to use in our model.

We'll do two loops:
1. The first will look at the correlation between the Permit Columns and the `hv_1yf` House Value column
2. The second will look at the correlation between the Permit Columns and the `pc_1yf` Percent Change column

The reason for the two loops is that the permit types that may have a higher correlation with the house value over the years may not have as high of a correlation with percent change, and vice versa.  Sussing those specific permit types out will be the goal of the below loops.

## Model Building

### Future Property Values - Selected Permits

We'll start with the first loop and build a model that predicts future property values, aka the `hv_1yf` column.

```python
housing_data.head()
```

```python
permit_type_list
```

```python
selected_permits = []

for permit_type in permit_type_list:
    permit_data = created_combined_df(permits_combined, [permit_type])
    
    # 12 is the hv_1yf column, 1:5 are the permit columns.
    correlation_mat = permit_data.corr().iloc[12, 1:5].round(2)
    sorted_pairs = correlation_mat.sort_values(kind="quicksort", ascending = False)
    # Select permit_type with a correlation >= 0.4
    if sorted_pairs[:1][0] >= 0.4:
        selected_permits.append(permit_type)
    print(permit_type)
    print(sorted_pairs[:1][0])
    print('-----')
```

The following permits had a correlation coefficient over 0.4 with the `hv_1yf` column indicating a moderate correlation.

```python
selected_permits
```

These selected permits will be used going forward in building the model that will predict house values a year into the future.  First let's look at the revised correlation matrix.

```python
hv_model_data = created_combined_df(permits_combined, selected_permits)
sns.heatmap(hv_model_data.corr().iloc[1:12, 12:13], annot = True)
plt.title("Correlation Matrix for Housing Data (Select Permit Types)")
plt.xlabel("House Value 1 Year Into The Future")
plt.ylabel("Permit Columns & House Value Columns (Past)")
plt.show()
```

The correlation in the upper section of the matrix showing the relationship between the permit columns and `hv_1yf` increased from a high of 0.16 when using all permits to 0.46.  We now have several columns with moderate to strong correlations that we can utilize to build our model.

Let's first build a model only using the house value columns, without any permit data.

```python
columns = [
    'hv_1yb',
    'hv_3yb',
    'hv_5yb'
]
```

```python
X = hv_model_data[columns]
y = hv_model_data['hv_1yf']
degree_list = [_ for _ in range(1,5)]

for d in degree_list:
    poly = PolynomialFeatures(degree=d, include_bias=False)
    poly_features = poly.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=1)
    poly_reg_model = LinearRegression()
    poly_reg_model.fit(X_train, y_train)
    
    # Predict values and compute RMSE
    test_predictions = poly_reg_model.predict(X_test)
    test_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))

    # Compute train RMSE to determine if model is overfitting the data at higher degrees
    train_predictions = poly_reg_model.predict(X_train)
    train_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))

    print("Degrees: ",d)
    print("Test RMSE: ",test_rmse)
    print("Train RMSE: ", train_rmse)
    print("-----")
```

The model with 3 degrees has a lower Test RMSE than the other models, however the difference between the Test RMSE and Train RMSE indicate that the model with 3 degrees is overfitting the data.  The model with 1 degree (linear regression) will be used going forward.

```python
X = hv_model_data[columns]
y = hv_model_data['hv_1yf']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)
predictions = lin_reg_model.predict(X_test)
rmse1 = np.sqrt(mean_squared_error(y_test, predictions))
print(rmse1)
```

Now let's add the permit columns to the model to determine if the model is more or less accurate.

```python
columns = [
    'no_permits',
    'np_1yb',
    'np_3yb',
    'np_5yb',
    'hv_1yb',
    'hv_3yb',
    'hv_5yb'
]
```

```python
X = hv_model_data[columns]
y = hv_model_data['hv_1yf']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)
predictions = lin_reg_model.predict(X_test)
rmse2 = np.sqrt(mean_squared_error(y_test, predictions))
print(rmse2)
```

```python
print(((rmse1-rmse2)/rmse1*100).round(2), "% reduced error")
```

Our model designed to predict future housing values with permit data had a 5.62% smaller error than the same model without permit data, which indicates that including permit data in a machine learning model may be beneficial in predicting future property values more accurately

Our goal is to minimize Test RMSE as well as the difference between Test and Train RMSE, so we'll again use Linear Regression as our model.
